# Import Library(PythaiNLP)

```python
from pythainlp import word_tokenize
```

```python
text = "ก็จะรู้ความชั่วร้ายที่ทำไว้    และคงจะไม่ยอมให้ทำนาบนหลังคน"
```

## Dictionary based

fast but low accuracy

```python
print(word_tokenize(text, engine="longest")) # logest matching
print(word_tokenize(text, engine="mm")) # maximal matching
print(word_tokenize(text, engine="newmm")) # maximal matching 2
print(word_tokenize(text, engine="newmm-safe")) # maximal matching 3
print(word_tokenize(text, engine="icu")) #dict icu(standard thai for translate thai to english)
```

## CNN Based

high accuracy but slow

```python
print(word_tokenize(text, engine="deepcut"))
print(word_tokenize(text, engine="attacut"))
```

## comparative time Dictionary based and CNN Based

```python
%%timeit
print(word_tokenize(text, engine="longest"))
```

```python
%%timeit
print(word_tokenize(text, engine="deepcut"))
```

```python
%%timeit
print(word_tokenize(text, engine="attacut"))
```

# Import Library(NECTEC)

```python
import requests
url_lextop = "https://api.aiforthai.in.th/lextoplus"
url_tiexp = "https://api.aiforthai.in.th/tiexplus"
headers = {'Apikey':""}
```

```python
def ai4thai_word_tokenize(url, header, text):
  params = {'text':text,'norm':'1'}
  res = requests.get(url, params=params, headers=headers).json()
  return res['tokens']
```

```python
text = "ก็จะรู้ความชั่วร้ายที่ทำไว้    และคงจะไม่ยอมให้ทำนาบนหลังคน"
print(ai4thai_word_tokenize(url_lextop, headers, text))
print(ai4thai_word_tokenize(url_tiexp, headers, text))
```
