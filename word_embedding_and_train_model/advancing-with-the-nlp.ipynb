# Word Embeddings

Word Embedding is a technique of word representation that allows words with similar meaning to be understood by machine learning algorithms. Technically speaking, it is a mapping of words into vectors of real numbers using the neural network, probabilistic model, or dimension reduction on word co-occurrence matrix.

Word embeddings are also very useful in mitigating the curse of dimensionality, a very recurring problem in artificial intelligence.

# Word2vec

It plots the words in a multi-dimensional vector space, where similar words tend to be close to each other. The surrounding words of a word provide the context to that word.

> Different architecture present in the **Word2Vec**

1. In the first option , CBOW; The model predicts the current word from a window of surrounding context words. The order of context words does not influence prediction (bag-of-words assumption)

2. Whereas, in the second option of using the continuous skip-gram architecture; the model uses the current word to predict the surrounding window of context words. The skip-gram architecture weighs nearby context words more heavily than more distant context words

```python
for i ,v in enumerate(w2v_vect):
    print(len(X_test.iloc[i]),len(v))
```

```python
w2v_vect_avg=[]

for vect in w2v_vect:
    if len(vect)!=0:
        w2v_vect_avg.append(vect.mean(axis=0))
    else:
        w2v_vect_avg.append(np.zeros(100))
```

```python
for i ,v in enumerate(w2v_vect_avg):
    print(len(X_test.iloc[i]),len(v))
```

# Doc2Vec

Doc2Vec is another widely used technique that creates an embedding of a document irrespective to its length. While Word2Vec computes a feature vector for every word in the corpus, Doc2Vec computes a feature vector for every document in the corpus. Doc2vec model is based on Word2Vec, with only adding another vector (paragraph ID) to the input. The Doc2Vec model, by analogy with Word2Vec, can rely on one of two model architectures which are: Distributed Memory version of Paragraph Vector (PV-DM) and Distributed Bag of Words version of Paragraph Vector (PV-DBOW).

## Keras for preprocessing

```python
import keras
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
#initializing and fitting the model in the keras tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
```

```python
#transform the train and the test dataset into the form of sequence
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)
```

```python
X_train_seq[0]
```

```python
#it is important to do the padding for our dataset becuase the keywords that we are using
#are of different lengths and before feeding it into the model we need to make sure , 
#that they all are of same lengths.
X_train_seq_padded = pad_sequences(X_train_seq,50)
X_test_seq_padded = pad_sequences(X_test_seq,50)
```

```python
X_train_seq_padded[0] #in this way we got all the arrays of the same length.
```

```python
#from here we will start by defining the model.
import keras.backend as K
from keras.layers import Dense , Embedding , LSTM
from keras.models import Sequential
```

```python
model = Sequential()

model.add(Embedding(len(tokenizer.index_word)+1,32))
model.add(LSTM(32,dropout=0,recurrent_dropout=0))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(X_train_seq_padded,y_train,batch_size=32,epochs=15)
```

```python
predicted_classes = model.predict(X_test_seq_padded)
```
